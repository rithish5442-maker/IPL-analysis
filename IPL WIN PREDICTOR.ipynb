```python
# ================================================================
#   IPL MATCH WINNER PREDICTION — COMPLETE JUPYTER NOTEBOOK
#   Copy each section into a separate Jupyter cell
#   Author: Your Name
# ================================================================


# ╔══════════════════════════════════════════╗
# ║  CELL 1 — IMPORTS                        ║
# ╚══════════════════════════════════════════╝

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')

print("✅ All libraries imported successfully")


# ╔══════════════════════════════════════════╗
# ║  CELL 2 — LOAD DATA                      ║
# ╚══════════════════════════════════════════╝

ipl        = pd.read_csv("D:\Downloads\ipl.csv")
deliveries = pd.read_csv("D:\\Downloads\\deliveries.csv")
batters    = pd.read_csv("D:\\Downloads\\IPL2025Batters.csv")

# Verify each file loaded correctly
print("ipl.csv columns:")
print(ipl.columns.tolist())

print("\ndeliveries.csv columns:")
print(deliveries.columns.tolist())

print("\nbatters.csv columns:")
print(batters.columns.tolist())
# ╔══════════════════════════════════════════╗
# ║  CELL 3 — EDA: IPL.CSV                   ║
# ╚══════════════════════════════════════════╝

print("=" * 50)
print("  EDA — ipl.csv")
print("=" * 50)

print("\n📋 First 5 rows:")
display(ipl.head())

print("\n📋 Column info:")
print(ipl.dtypes)

print("\n📋 Null values:")
print(ipl.isnull().sum())

print("\n📋 Result types:")
print(ipl['result'].value_counts())

print("\n📋 Toss decisions:")
print(ipl['toss_decision'].value_counts())

print(f"\n📋 Seasons in data: {sorted(ipl['season'].unique())}")



# ╔══════════════════════════════════════════╗
# ║  CELL 4 — EDA: DELIVERIES.CSV            ║
# ╚══════════════════════════════════════════╝

print("=" * 50)
print("  EDA — deliveries.csv")
print("=" * 50)

print("\n📋 First 5 rows:")
display(deliveries.head())

print(f"\n📋 Over range      : {deliveries['over'].min()} to {deliveries['over'].max()}")
print(f"📋 Unique matches  : {deliveries['match_id'].nunique()}")
print(f"📋 Unique teams    : {deliveries['batting_team'].nunique()}")
print(f"📋 Total wickets   : {deliveries['is_wicket'].sum():,}")
print(f"📋 Total runs      : {deliveries['total_runs'].sum():,}")


# ╔══════════════════════════════════════════╗
# ║  CELL 5 — EDA: BATTERS.CSV               ║
# ╚══════════════════════════════════════════╝

print("=" * 50)
print("  EDA — IPL2025Batters.csv")
print("=" * 50)

print("\n📋 First 5 rows:")
display(batters.head())

print("\n📋 Top 5 run scorers:")
display(batters.sort_values('Runs', ascending=False).head())

print("\n📋 Top 5 by Strike Rate (min 100 runs):")
display(batters[batters['Runs'] >= 100].sort_values('SR', ascending=False).head())

print("\n📋 Teams represented:")
print(batters['Team'].value_counts())

import os

# Create outputs folder if it doesn't exist
os.makedirs('outputs', exist_ok=True)
print("✅ outputs/ folder created")

# ╔══════════════════════════════════════════╗
# ║  CELL 6 — EDA VISUALISATIONS             ║
# ╚══════════════════════════════════════════╝

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('IPL Data — Exploratory Analysis', fontsize=15, fontweight='bold')
COLORS = ['#2196F3','#FF9800','#4CAF50','#E91E63','#9C27B0','#00BCD4','#FF5722','#607D8B']

# ── Matches per season ──────────────────────────────────
ax1 = axes[0, 0]
matches_per_season = ipl.groupby('season').size()
ax1.bar(matches_per_season.index, matches_per_season.values, color=COLORS[0])
ax1.set_title('Matches per Season', fontweight='bold')
ax1.set_xlabel('Season'); ax1.set_ylabel('Matches')

# ── Toss decision distribution ──────────────────────────
ax2 = axes[0, 1]
toss = ipl['toss_decision'].value_counts()
ax2.pie(toss.values, labels=toss.index, autopct='%1.1f%%',
        colors=[COLORS[0], COLORS[1]], startangle=90)
ax2.set_title('Toss Decision Distribution', fontweight='bold')

# ── Top 10 run scorers (2025) ────────────────────────────
ax3 = axes[1, 0]
top10 = batters.sort_values('Runs', ascending=True).tail(10)
ax3.barh(top10['Player Name'], top10['Runs'],
         color=[COLORS[i % len(COLORS)] for i in range(10)])
ax3.set_title('Top 10 Run Scorers — IPL 2025', fontweight='bold')
ax3.set_xlabel('Runs')

# ── Top teams by wins ───────────────────────────────────
ax4 = axes[1, 1]
top_winners = ipl['winner'].value_counts().head(8)
ax4.bar(top_winners.index, top_winners.values,
        color=[COLORS[i % len(COLORS)] for i in range(8)])
ax4.set_title('All-Time Wins by Team', fontweight='bold')
ax4.set_ylabel('Wins')
ax4.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('outputs/eda_overview.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ EDA chart saved to outputs/eda_overview.png")


# ╔══════════════════════════════════════════╗
# ║  CELL 7 — DATA CLEANING                  ║
# ╚══════════════════════════════════════════╝

print("=" * 50)
print("  DATA CLEANING")
print("=" * 50)

# Step 1: Keep only valid matches
print(f"\nBefore cleaning : {len(ipl)} matches")
ipl = ipl[ipl['result'].isin(['runs', 'wickets', 'tie'])].copy()
ipl = ipl[ipl['winner'].notna()].copy()
print(f"After removing no result/abandoned : {len(ipl)} matches")

# Step 2: Remove super over matches
ipl = ipl[ipl['super_over'] == 'N'].copy()
print(f"After removing super overs : {len(ipl)} matches")

# Step 3: Map season number to year
season_map = {i: 2007 + i for i in range(1, 19)}
ipl['year'] = ipl['season'].map(season_map)
print(f"\nSeason year range: {ipl['year'].min()} to {ipl['year'].max()}")

# Step 4: Standardise team names
name_fix = {
    'Delhi Daredevils'            : 'Delhi Capitals',
    'Kings XI Punjab'             : 'Punjab Kings',
    'Rising Pune Supergiant'      : 'Rising Pune Supergiants',
    'Royal Challengers Bangalore' : 'Royal Challengers Bengaluru'
}
for col in ['team1', 'team2', 'winner', 'toss_winner']:
    ipl[col] = ipl[col].replace(name_fix)
for col in ['batting_team', 'bowling_team']:
    deliveries[col] = deliveries[col].replace(name_fix)

ipl = ipl.reset_index(drop=True)

# Verify
print("\n✅ Old team names remaining:")
for name in name_fix.keys():
    count = ((ipl['team1']==name)|(ipl['team2']==name)).sum()
    print(f"   '{name}' → {count} (should be 0)")

print(f"\n✅ Final clean dataset: {len(ipl)} matches")
print(f"✅ Any nulls in winner: {ipl['winner'].isnull().sum()}")


# ╔══════════════════════════════════════════╗
# ║  CELL 8 — DELIVERY FEATURES              ║
# ╚══════════════════════════════════════════╝

print("=" * 50)
print("  FEATURE ENGINEERING — DELIVERY STATS")
print("=" * 50)

# ── Powerplay (overs 0–5) ────────────────────────────────
pp = deliveries[deliveries['over'] <= 5].groupby('batting_team').agg(
    pp_runs    = ('total_runs', 'sum'),
    pp_balls   = ('total_runs', 'count'),
    pp_wickets = ('is_wicket',  'sum')
).reset_index()
pp['pp_run_rate']    = pp['pp_runs']    / (pp['pp_balls'] / 6)
pp['pp_wicket_rate'] = pp['pp_wickets'] /  pp['pp_balls'] * 6

# ── Death overs (overs 16–19) ────────────────────────────
death = deliveries[deliveries['over'] >= 16].groupby('batting_team').agg(
    death_runs    = ('total_runs', 'sum'),
    death_balls   = ('total_runs', 'count'),
    death_wickets = ('is_wicket',  'sum')
).reset_index()
death['death_run_rate']    = death['death_runs']    / (death['death_balls'] / 6)
death['death_wicket_rate'] = death['death_wickets'] /  death['death_balls'] * 6

# ── Bowling economy ──────────────────────────────────────
bowl = deliveries.groupby('bowling_team').agg(
    bowl_runs  = ('total_runs', 'sum'),
    bowl_balls = ('total_runs', 'count'),
    bowl_wkts  = ('is_wicket',  'sum')
).reset_index()
bowl['bowl_economy']     = bowl['bowl_runs']  / (bowl['bowl_balls'] / 6)
bowl['bowl_wicket_rate'] = bowl['bowl_wkts']  /  bowl['bowl_balls'] * 6

# ── Merge all delivery features ──────────────────────────
team_features = (
    pp[['batting_team', 'pp_run_rate', 'pp_wicket_rate']]
    .merge(death[['batting_team', 'death_run_rate', 'death_wicket_rate']], on='batting_team')
    .merge(bowl[['bowling_team', 'bowl_economy', 'bowl_wicket_rate']]
           .rename(columns={'bowling_team':'batting_team',
                            'bowl_wicket_rate':'bowl_wicket_rate_against'}),
           on='batting_team', how='left')
)

print("\n✅ Delivery features computed:")
display(team_features.sort_values('pp_run_rate', ascending=False).head(5))


# ╔══════════════════════════════════════════╗
# ║  CELL 9 — PLAYER STRENGTH SCORE          ║
# ╚══════════════════════════════════════════╝

print("=" * 50)
print("  FEATURE ENGINEERING — PLAYER STRENGTH")
print("=" * 50)

# Map short codes to full team names
abbrev_map = {
    'GT' : 'Gujarat Titans',
    'MI' : 'Mumbai Indians',
    'RCB': 'Royal Challengers Bengaluru',
    'LSG': 'Lucknow Super Giants',
    'PBKS':'Punjab Kings',
    'RR' : 'Rajasthan Royals',
    'DC' : 'Delhi Capitals',
    'SRH': 'Sunrisers Hyderabad',
    'KKR': 'Kolkata Knight Riders',
    'CSK': 'Chennai Super Kings'
}
batters['full_team'] = batters['Team'].map(abbrev_map)

# Fix AVG column (stored as string)
batters['AVG'] = pd.to_numeric(batters['AVG'], errors='coerce').fillna(0)

# Composite strength score
batters['strength_score'] = (
    0.5 * batters['AVG']         +   # Consistency
    0.3 * (batters['SR'] / 10)   +   # Aggression
    0.1 * batters['100s'] * 10   +   # Match-winning ability
    0.1 * batters['50s']  * 3        # Contribution frequency
)

team_strength = batters.groupby('full_team')['strength_score'].mean().reset_index()
team_strength.columns = ['team', 'batting_strength']

print("\n✅ Batting strength scores:")
display(team_strength.sort_values('batting_strength', ascending=False))


# ╔══════════════════════════════════════════╗
# ║  CELL 10 — HELPER FUNCTIONS              ║
# ╚══════════════════════════════════════════╝

def get_team_delivery_stats(team):
    """Fetch delivery-based stats for a team."""
    row = team_features[team_features['batting_team'] == team]
    if len(row) == 0:
        return {
            'pp_run_rate': 7.5, 'pp_wicket_rate': 0.5,
            'death_run_rate': 9.5, 'death_wicket_rate': 0.8,
            'bowl_economy': 8.0, 'bowl_wicket_rate_against': 0.5
        }
    return row.iloc[0].to_dict()

def get_batting_strength(team):
    """Fetch batting strength score for a team."""
    row = team_strength[team_strength['team'] == team]
    return row['batting_strength'].values[0] if len(row) > 0 else 25.0

print("✅ Helper functions defined")


# ╔══════════════════════════════════════════╗
# ║  CELL 11 — FEATURE ENGINEERING PER MATCH ║
# ╚══════════════════════════════════════════╝

print("=" * 50)
print("  BUILDING MATCH-LEVEL FEATURES")
print("  (No data leakage — uses only past matches)")
print("=" * 50)

features = []

for idx, row in ipl.iterrows():
    t1, t2   = row['team1'], row['team2']
    venue    = row['venue']
    toss_w   = row['toss_winner']
    toss_d   = row['toss_decision']
    year     = row['year']

    # Only use matches BEFORE current match
    past = ipl.iloc[:idx]

    # ── Historical win rates ──────────────────────────────
    t1_matches  = past[(past['team1']==t1) | (past['team2']==t1)]
    t1_win_rate = (past['winner']==t1).sum() / len(t1_matches) if len(t1_matches)>0 else 0.5

    t2_matches  = past[(past['team1']==t2) | (past['team2']==t2)]
    t2_win_rate = (past['winner']==t2).sum() / len(t2_matches) if len(t2_matches)>0 else 0.5

    # ── Head-to-head win rate ─────────────────────────────
    h2h = past[
        ((past['team1']==t1) & (past['team2']==t2)) |
        ((past['team1']==t2) & (past['team2']==t1))
    ]
    h2h_win_rate = (h2h['winner']==t1).sum() / len(h2h) if len(h2h)>0 else 0.5

    # ── Venue win rate ────────────────────────────────────
    venue_matches = past[(past['venue']==venue) & ((past['team1']==t1)|(past['team2']==t1))]
    venue_wins    = past[(past['venue']==venue) & (past['winner']==t1)].shape[0]
    venue_rate    = venue_wins / len(venue_matches) if len(venue_matches)>0 else 0.5

    # ── Recent form (last 5 matches) ──────────────────────
    t1_form = (t1_matches.tail(5)['winner']==t1).sum() / min(5, max(1, len(t1_matches)))
    t2_form = (t2_matches.tail(5)['winner']==t2).sum() / min(5, max(1, len(t2_matches)))

    # ── Delivery-based features ───────────────────────────
    t1d = get_team_delivery_stats(t1)
    t2d = get_team_delivery_stats(t2)

    # ── Player strength ───────────────────────────────────
    t1s = get_batting_strength(t1)
    t2s = get_batting_strength(t2)

    # ── Target label ──────────────────────────────────────
    label = 1 if row['winner'] == t1 else 0

    features.append({
        # Historical
        'team1_win_rate'         : t1_win_rate,
        'team2_win_rate'         : t2_win_rate,
        'win_rate_diff'          : t1_win_rate - t2_win_rate,
        'h2h_win_rate_team1'     : h2h_win_rate,
        'venue_win_rate_team1'   : venue_rate,
        # Toss
        'toss_winner_is_team1'   : 1 if toss_w==t1 else 0,
        'toss_chose_bat'         : 1 if toss_d=='bat' else 0,
        # Form
        'team1_recent_form'      : t1_form,
        'team2_recent_form'      : t2_form,
        'form_diff'              : t1_form - t2_form,
        # Powerplay
        'pp_rr_diff'             : t1d['pp_run_rate']    - t2d['pp_run_rate'],
        'pp_wkt_diff'            : t1d['pp_wicket_rate'] - t2d['pp_wicket_rate'],
        # Death overs
        'death_rr_diff'          : t1d['death_run_rate']    - t2d['death_run_rate'],
        'death_wkt_diff'         : t1d['death_wicket_rate'] - t2d['death_wicket_rate'],
        # Bowling
        'bowl_eco_diff'          : t1d['bowl_economy'] - t2d['bowl_economy'],
        # Player strength
        'batting_strength_diff'  : t1s - t2s,
        'team1_batting_strength' : t1s,
        'team2_batting_strength' : t2s,
        # Season
        'season'                 : year,
        'label'                  : label
    })

df = pd.DataFrame(features)
FEATURE_COLS = [c for c in df.columns if c != 'label']

print(f"\n✅ Feature matrix built: {df.shape[0]} rows × {len(FEATURE_COLS)} features")
print(f"✅ Any NaN: {df.isnull().sum().sum()}")
print(f"✅ Label balance: {df['label'].mean()*100:.1f}% team1 wins")


# ╔══════════════════════════════════════════╗
# ║  CELL 12 — TRAIN / TEST SPLIT            ║
# ╚══════════════════════════════════════════╝

print("=" * 50)
print("  TRAIN / TEST SPLIT")
print("=" * 50)

X = df[FEATURE_COLS]
y = df['label']

# Chronological split — no data leakage
split_year  = 2020
train_mask  = df['season'] < split_year
X_train, y_train = X[train_mask],  y[train_mask]
X_test,  y_test  = X[~train_mask], y[~train_mask]

print(f"\n✅ Training set  : {len(X_train)} matches (2008–2019)")
print(f"✅ Testing set   : {len(X_test)} matches (2020–2025)")
print(f"✅ Train ratio   : {len(X_train)/len(X)*100:.1f}%")
print(f"✅ Season overlap: {set(df[train_mask]['season']) & set(df[~train_mask]['season'])} (should be empty)")
print(f"\n✅ Train label balance: {y_train.mean()*100:.1f}% team1 wins")
print(f"✅ Test  label balance: {y_test.mean()*100:.1f}% team1 wins")


# ╔══════════════════════════════════════════╗
# ║  CELL 13 — TRAIN MODELS                  ║
# ╚══════════════════════════════════════════╝

print("=" * 50)
print("  TRAINING MODELS")
print("=" * 50)

# ── Gradient Boosting ─────────────────────────────────────
print("\nTraining Gradient Boosting...")
gb = GradientBoostingClassifier(
    n_estimators   = 100,
    max_depth      = 2,
    learning_rate  = 0.01,
    subsample      = 0.6,
    min_samples_leaf = 10,
    random_state   = 42
)
gb.fit(X_train, y_train)
gb_train_acc = accuracy_score(y_train, gb.predict(X_train))
gb_test_acc  = accuracy_score(y_test,  gb.predict(X_test))
print(f"   Train: {gb_train_acc*100:.1f}%  |  Test: {gb_test_acc*100:.1f}%  |  Gap: {abs(gb_train_acc-gb_test_acc)*100:.1f}%")

# ── Random Forest ─────────────────────────────────────────
print("\nTraining Random Forest...")
rf = RandomForestClassifier(
    n_estimators   = 200,
    max_depth      = 3,
    min_samples_leaf = 30,
    max_features   = 'sqrt',
    random_state   = 42
)
rf.fit(X_train, y_train)
rf_train_acc = accuracy_score(y_train, rf.predict(X_train))
rf_test_acc  = accuracy_score(y_test,  rf.predict(X_test))
print(f"   Train: {rf_train_acc*100:.1f}%  |  Test: {rf_test_acc*100:.1f}%  |  Gap: {abs(rf_train_acc-rf_test_acc)*100:.1f}%")

# ── Pick best model ───────────────────────────────────────
best_model      = gb if gb_test_acc >= rf_test_acc else rf
best_model_name = 'Gradient Boosting' if gb_test_acc >= rf_test_acc else 'Random Forest'
best_preds      = gb.predict(X_test) if gb_test_acc >= rf_test_acc else rf.predict(X_test)

print(f"\n✅ Best model: {best_model_name} ({max(gb_test_acc, rf_test_acc)*100:.1f}% test accuracy)")


# ╔══════════════════════════════════════════╗
# ║  CELL 14 — EVALUATION                    ║
# ╚══════════════════════════════════════════╝

print("=" * 50)
print("  MODEL EVALUATION")
print("=" * 50)

# ── Classification report ─────────────────────────────────
print(f"\n📊 Classification Report ({best_model_name}):")
print(classification_report(y_test, best_preds,
      target_names=['Team2 Wins', 'Team1 Wins']))

# ── Cross validation ──────────────────────────────────────
print("📊 5-Fold Cross Validation:")
gb_cv = cross_val_score(gb, X, y, cv=5, scoring='accuracy')
rf_cv = cross_val_score(rf, X, y, cv=5, scoring='accuracy')
print(f"   GB  → Mean: {gb_cv.mean()*100:.1f}%  Std: {gb_cv.std()*100:.2f}%")
print(f"   RF  → Mean: {rf_cv.mean()*100:.1f}%  Std: {rf_cv.std()*100:.2f}%")

# ── Overfitting check ─────────────────────────────────────
print("\n📊 Overfitting Check:")
print(f"   GB  gap: {abs(gb_train_acc-gb_test_acc)*100:.1f}%  → {'✅ Acceptable' if abs(gb_train_acc-gb_test_acc)<0.15 else '⚠️ High'}")
print(f"   RF  gap: {abs(rf_train_acc-rf_test_acc)*100:.1f}%  → {'✅ Acceptable' if abs(rf_train_acc-rf_test_acc)<0.15 else '⚠️ High'}")

# ── Per-season accuracy ───────────────────────────────────
print("\n📊 Per-Season Test Accuracy:")
test_df = df[~train_mask].copy()
test_df['gb_pred'] = gb.predict(X_test)
test_df['rf_pred'] = rf.predict(X_test)
for yr in sorted(test_df['season'].unique()):
    sub = test_df[test_df['season'] == yr]
    gb_yr = accuracy_score(sub['label'], sub['gb_pred'])
    rf_yr = accuracy_score(sub['label'], sub['rf_pred'])
    print(f"   {int(yr)} ({len(sub):3} matches) — GB: {gb_yr*100:.1f}%  RF: {rf_yr*100:.1f}%")


# ╔══════════════════════════════════════════╗
# ║  CELL 15 — VISUALISATIONS                ║
# ╚══════════════════════════════════════════╝

PALETTE = ['#2196F3','#FF9800','#4CAF50','#E91E63','#9C27B0','#00BCD4','#FF5722','#607D8B']

fig = plt.figure(figsize=(16, 14))
fig.suptitle('IPL Match Winner Prediction — Final Results',
             fontsize=17, fontweight='bold', y=0.99)
gs = gridspec.GridSpec(3, 3, figure=fig, hspace=0.5, wspace=0.4)

# ── 1. Model Accuracy Comparison ──────────────────────────
ax1 = fig.add_subplot(gs[0, 0])
models = ['Gradient\nBoosting', 'Random\nForest']
accs   = [gb_test_acc * 100, rf_test_acc * 100]
bars   = ax1.bar(models, accs, color=[PALETTE[0], PALETTE[1]], width=0.45)
ax1.set_ylim(40, 70)
ax1.set_title('Test Accuracy Comparison', fontweight='bold')
ax1.set_ylabel('Accuracy %')
ax1.axhline(50, color='red', linestyle='--', linewidth=1, label='Baseline')
ax1.legend(fontsize=8)
for b, v in zip(bars, accs):
    ax1.text(b.get_x()+b.get_width()/2, v+0.5, f'{v:.1f}%',
             ha='center', fontweight='bold', fontsize=11)

# ── 2. Train vs Test Gap ───────────────────────────────────
ax2 = fig.add_subplot(gs[0, 1])
labels = ['GB Train','GB Test','RF Train','RF Test']
values = [gb_train_acc*100, gb_test_acc*100, rf_train_acc*100, rf_test_acc*100]
colors = [PALETTE[0], PALETTE[2], PALETTE[1], PALETTE[3]]
bars2  = ax2.bar(labels, values, color=colors, width=0.5)
ax2.set_ylim(40, 80)
ax2.set_title('Train vs Test Accuracy', fontweight='bold')
ax2.set_ylabel('Accuracy %')
for b, v in zip(bars2, values):
    ax2.text(b.get_x()+b.get_width()/2, v+0.5, f'{v:.1f}%',
             ha='center', fontsize=9, fontweight='bold')

# ── 3. Cross Validation ────────────────────────────────────
ax3 = fig.add_subplot(gs[0, 2])
folds = ['F1','F2','F3','F4','F5']
ax3.plot(folds, gb_cv*100, 'o-', color=PALETTE[0], label='GB', linewidth=2, markersize=8)
ax3.plot(folds, rf_cv*100, 's-', color=PALETTE[1], label='RF', linewidth=2, markersize=8)
ax3.axhline(50, color='red', linestyle='--', linewidth=1, label='Baseline')
ax3.set_title('5-Fold Cross Validation', fontweight='bold')
ax3.set_ylabel('Accuracy %')
ax3.set_ylim(40, 65)
ax3.legend(fontsize=8)

# ── 4. Feature Importance ──────────────────────────────────
ax4 = fig.add_subplot(gs[1, :2])
imp  = pd.Series(best_model.feature_importances_, index=FEATURE_COLS).sort_values(ascending=True).tail(12)
clrs = [PALETTE[i % len(PALETTE)] for i in range(len(imp))]
bars4 = ax4.barh(imp.index, imp.values, color=clrs)
ax4.set_title(f'Feature Importance ({best_model_name})', fontweight='bold')
ax4.set_xlabel('Score')
for b, v in zip(bars4, imp.values):
    ax4.text(v+0.001, b.get_y()+b.get_height()/2, f'{v:.3f}', va='center', fontsize=8)

# ── 5. Confusion Matrix ────────────────────────────────────
ax5 = fig.add_subplot(gs[1, 2])
cm = confusion_matrix(y_test, best_preds)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax5,
            xticklabels=['Team2 Wins','Team1 Wins'],
            yticklabels=['Team2 Wins','Team1 Wins'])
ax5.set_title('Confusion Matrix', fontweight='bold')
ax5.set_ylabel('Actual'); ax5.set_xlabel('Predicted')

# ── 6. Powerplay Run Rate ──────────────────────────────────
ax6 = fig.add_subplot(gs[2, :])
top_teams = ['Mumbai Indians','Chennai Super Kings','Kolkata Knight Riders',
             'Royal Challengers Bengaluru','Delhi Capitals','Rajasthan Royals',
             'Sunrisers Hyderabad','Punjab Kings','Gujarat Titans','Lucknow Super Giants']
pp_plot = team_features[team_features['batting_team'].isin(top_teams)][['batting_team','pp_run_rate']].sort_values('pp_run_rate')
clrs6   = [PALETTE[i % len(PALETTE)] for i in range(len(pp_plot))]
bars6   = ax6.barh(pp_plot['batting_team'], pp_plot['pp_run_rate'], color=clrs6)
ax6.set_title('Powerplay Run Rate by Team (All-time)', fontweight='bold')
ax6.set_xlabel('Run Rate (Overs 1–6)')
for b, v in zip(bars6, pp_plot['pp_run_rate'].values):
    ax6.text(v+0.02, b.get_y()+b.get_height()/2, f'{v:.2f}', va='center', fontsize=9)

plt.savefig('outputs/model_results.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart saved to outputs/model_results.png")


# ╔══════════════════════════════════════════╗
# ║  CELL 16 — PREDICTION FUNCTION           ║
# ╚══════════════════════════════════════════╝

def predict_match(team1, team2, venue, toss_winner, toss_decision):
    """
    Predict which team wins an IPL match before it starts.

    Parameters:
    -----------
    team1         : str  — First team name
    team2         : str  — Second team name
    venue         : str  — Stadium name
    toss_winner   : str  — Who won the toss (team1 or team2)
    toss_decision : str  — 'bat' or 'field'
    """
    past = ipl.copy()

    t1m  = past[(past['team1']==team1)|(past['team2']==team1)]
    t2m  = past[(past['team1']==team2)|(past['team2']==team2)]
    t1wr = (past['winner']==team1).sum()/len(t1m) if len(t1m)>0 else 0.5
    t2wr = (past['winner']==team2).sum()/len(t2m) if len(t2m)>0 else 0.5

    h2h  = past[((past['team1']==team1)&(past['team2']==team2))|
                ((past['team1']==team2)&(past['team2']==team1))]
    h2hw = (h2h['winner']==team1).sum()/len(h2h) if len(h2h)>0 else 0.5

    vm   = past[(past['venue']==venue)&((past['team1']==team1)|(past['team2']==team1))]
    vw   = past[(past['venue']==venue)&(past['winner']==team1)].shape[0]
    vwr  = vw/len(vm) if len(vm)>0 else 0.5

    t1f  = (t1m.tail(5)['winner']==team1).sum()/min(5,max(1,len(t1m)))
    t2f  = (t2m.tail(5)['winner']==team2).sum()/min(5,max(1,len(t2m)))
    t1d,t2d = get_team_delivery_stats(team1), get_team_delivery_stats(team2)
    t1s,t2s = get_batting_strength(team1), get_batting_strength(team2)

    X_new = pd.DataFrame([{
        'team1_win_rate'         : t1wr,
        'team2_win_rate'         : t2wr,
        'win_rate_diff'          : t1wr - t2wr,
        'h2h_win_rate_team1'     : h2hw,
        'venue_win_rate_team1'   : vwr,
        'toss_winner_is_team1'   : 1 if toss_winner==team1 else 0,
        'toss_chose_bat'         : 1 if toss_decision=='bat' else 0,
        'team1_recent_form'      : t1f,
        'team2_recent_form'      : t2f,
        'form_diff'              : t1f - t2f,
        'pp_rr_diff'             : t1d['pp_run_rate']    - t2d['pp_run_rate'],
        'pp_wkt_diff'            : t1d['pp_wicket_rate'] - t2d['pp_wicket_rate'],
        'death_rr_diff'          : t1d['death_run_rate']    - t2d['death_run_rate'],
        'death_wkt_diff'         : t1d['death_wicket_rate'] - t2d['death_wicket_rate'],
        'bowl_eco_diff'          : t1d['bowl_economy'] - t2d['bowl_economy'],
        'batting_strength_diff'  : t1s - t2s,
        'team1_batting_strength' : t1s,
        'team2_batting_strength' : t2s,
        'season'                 : 2026
    }])

    prob   = best_model.predict_proba(X_new)[0]
    winner = team1 if prob[1] > 0.5 else team2
    conf   = max(prob) * 100
    t1p    = prob[1] * 100
    t2p    = prob[0] * 100

    # ── Visual output ──────────────────────────────────────
    fig, ax = plt.subplots(figsize=(8, 3))
    ax.axis('off')
    ax.barh([team2, team1], [t2p, t1p],
            color=['#FF9800' if team2==winner else '#B0BEC5',
                   '#2196F3' if team1==winner else '#B0BEC5'],
            height=0.4)
    ax.set_xlim(0, 100)
    ax.axvline(50, color='gray', linestyle='--', linewidth=1)
    ax.text(t1p + 1, 1, f'{t1p:.1f}%', va='center', fontweight='bold')
    ax.text(t2p + 1, 0, f'{t2p:.1f}%', va='center', fontweight='bold')
    ax.set_title(f'🏏  {team1}  vs  {team2}\n'
                 f'Venue: {venue}  |  Toss: {toss_winner} chose to {toss_decision}\n'
                 f'🏆 Predicted Winner: {winner}  (Confidence: {conf:.1f}%)',
                 fontsize=11, fontweight='bold', pad=12)
    plt.tight_layout()
    plt.show()

    return winner, conf

print("✅ predict_match() function defined")


# ╔══════════════════════════════════════════╗
# ║  CELL 17 — RUN PREDICTIONS               ║
# ╚══════════════════════════════════════════╝

# ── Try your own matches here! ─────────────────────────────
predict_match(
    team1         = 'Mumbai Indians',
    team2         = 'Chennai Super Kings',
    venue         = 'Wankhede Stadium',
    toss_winner   = 'Mumbai Indians',
    toss_decision = 'bat'
)

predict_match(
    team1         = 'Royal Challengers Bengaluru',
    team2         = 'Kolkata Knight Riders',
    venue         = 'M Chinnaswamy Stadium',
    toss_winner   = 'Kolkata Knight Riders',
    toss_decision = 'field'
)

predict_match(
    team1         = 'Rajasthan Royals',
    team2         = 'Gujarat Titans',
    venue         = 'Sawai Mansingh Stadium',
    toss_winner   = 'Rajasthan Royals',
    toss_decision = 'field'
)
```
